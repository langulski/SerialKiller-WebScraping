### Web-scrapping - serial killers

This project is intend to compare the data from the surface of the internet and the statistics given from the Radford FGCU database, and check if the information is reliable. 

   Radford FGCU has the largest non-governmental database of serial killers in the world. Unfortunatly the dataset is not public, though it can be obtained through a request to the Florida Gulf Coast University (FGCU) and can be accessible for research.

```python
#Returning a new column as integer 
age_list = []
for row in sk_db['Ages Active:']:
    try:
        age_list.append(re.search("^.{0,2}\d+",str(row)).group())        
    except:
        
        age_list.append(row)    
        
sk_db['Starting Age'] = age_list
```
